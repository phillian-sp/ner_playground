```python
import pandas as pd
import json

with open("parent-utterances_annotations.json", "r", encoding="utf-8") as file:
    df1 = json.load(file)

df1
```

```python
all_annotations = []

for i in range(len(df1['examples'])):
    example = df1['examples'][i]
    sentence = example['content']
    sentence_id = example['example_id']  # Assuming 'id' is the sentence identifier in your JSON data
    
    for f in example['annotations']:
        if f['correct'] == False or f['reviewed'] == False:
            continue
        
        word = f['value']
        label = f['tag']
        start = f['start']  # Assuming 'start' is the start position of the word
        end = f['end']      # Assuming 'end' is the end position of the word
        
        d = {
            'sentence': sentence,
            'sentence_id': sentence_id,
            'word': word,
            'label': label,
            'start': start,
            'end': end
        }
        
        all_annotations.append(d)
```

```python
all_annotations
```

```python
with open("validation_parents.json", 'w', encoding='utf-8') as json_file:
    json.dump(all_annotations, json_file, ensure_ascii=False, indent=4)
```
